```python
import matplotlib.pyplot as plt
from scipy.io import wavfile

def visualize_waveform(wav_file):
    # Read the WAV file
    sample_rate, data = wavfile.read(wav_file)

    print(sample_rate, data)
    print(f"min: {data.min()}\nmax: {data.max()}\nmean: {data.mean()}")
    # Create a time array in seconds
    time = range(0, len(data))
    time = [t / sample_rate for t in time]

    # Plotting the waveform
    plt.figure(figsize=(12, 4))
    plt.plot(time, data)
    plt.title('Waveform of ' + wav_file)
    plt.xlabel('Time [s]')
    plt.ylabel('Amplitude')
    plt.show()

vocal_audio = "D:/Casper/Data/Taiwanese/wav/ZC04/train/ZC04-F01/ZC04-F01_01.wav"
sound_audio = "D:/Casper/TaiwaneseLM/multimodal/ImageBind/.assets/dog_audio.wav"
visualize_waveform(sound_audio)
```
